```python
dummies=pd.get_dummies(data[["Sex","Country","Continent","Hemisphere","Diet"]])
dummies
```

```python
dummi_droped=dummies.drop(["Sex_Female","Country_Argentina","Continent_Australia","Hemisphere_Northern Hemisphere","Diet_Average"],axis=1)
```

```python
data_droped=data.drop(["Sex","Country","Continent","Hemisphere","Diet"],axis=1)
```

```python
final_merged=pd.concat([data_droped,dummi_droped],axis=1)
final_merged
```

```python
labels=final_merged["Heart Attack Risk"].copy()
labels
```

```python
final_merged=final_merged.drop(["Patient ID","Blood Pressure"],axis=1)
```

```python
final_merged.info()
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scaled_final=scalar.fit_transform(final_merged)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=35)
reduced_final=pca.fit_transform(scaled_final)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(reduced_final ,labels ,test_size=.35 ,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
LR_model=LogisticRegression().fit(X_train,y_train)
pre=LR_model.predict(X_test)
```

```python
LR_model.score(X_train,y_train)
```

```python
LR_model.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pre))
```

```python
sample_data = X_test[:1]
sample_label=y_test[:1]
sample_predicted=LR_model.predict(sample_data)
print(sample_label)
print(sample_predicted)
```
