```python
dummies=pd.get_dummies(data[["Sex","Country","Continent","Hemisphere"]])
dummies
```

```python
dummi_droped=dummies.drop(["Sex_Female","Country_Argentina","Continent_Australia","Hemisphere_Northern Hemisphere"],axis=1)
```

```python
data_droped=data.drop(["Sex","Country","Continent","Hemisphere"],axis=1)
```

```python
final_merged=pd.concat([data_droped,dummi_droped],axis=1)
final_merged
```

```python
labels=final_merged["Heart Attack Risk"].copy()
labels
```

```python
final_merged=final_merged.drop(["Patient ID"],axis=1)
```

```python
final_merged.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(final_merged ,labels ,test_size=.35 ,random_state=101)
```
